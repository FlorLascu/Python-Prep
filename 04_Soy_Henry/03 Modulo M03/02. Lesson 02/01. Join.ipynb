#### Teoria de Conjuntos
##### Diagrama de Venn

- Esquemas utilizados en la teoria de Conjuntos<br>
- Muestran colecciones (Conjuntos) de cosas (elementos) por medio de lineas cerradas<br>
- La linea cerrada exterior abarca a todos los elementos bajo consideracion, el conjunto Universal U.<br>
- Usa circulos que se superponen u otras figuras que sirven para ilustrar las **relaciones logicas** entre dos o mas conjuntos de elementos<br>

##### Union de Conjuntos

Contiene cada elemento que est contenido, al menos, en alguno de ellos.<br>

###### A OR B = A U B

##### Insterseccion de Conjuntos

Incluye todos los elementos que estos conjuntos comparten o que tienen en comn.<br>

###### A AND B = A interseccion B

##### Diferencia de Conjuntos

Es igual a los elementos del primer conjunto menos los elementos del segundo.<br>


###### A - B = son todos los elementos de A menos los elementos que tambien estan en B (A - (A INTERSECCION B))

##### Conjuntos complementarios

Incluye todos los elementos que no estn contenidos en dicho conjunto (pero que s pertenecen a otro conjunto de referencia). <br>


###### NOT A  = SON TODOS LOS ELEMENTO DE U MENOS LOS ELEMENTOS DE A


##### Diferencia Simetrica

Incluye todos los elementos que estn en uno o en otro, pero no en ambos al mismo tiempo.<br>

##### Producto Cartesiano

Es una operacin que da como resultado un nuevo conjunto.<br>



#### JOIN
La instruccin **Join**, se usa para acceder los datos de dos tablas relacionadas a travs de algn campo en comn, este acceso se da gracias a las **foreign key** que permiten generar las relaciones entre ellas. Asimismo, permite tener un alto grado de normalizacin en las tablas y an as poder accederlas de forma sencilla.<br>

##### INNER JOIN
Devuelve solo aquellas filas donde haya un match<br>




##### LEFT JOIN
Cundo quieres todas las filas para las que haya match pero  tambin aquellas de la Tabla A que no hagan match.<br>




##### RIGHT JOIN
Exactamente lo mismo pero a la inversa, cuando quieres  listar las filas de la tabla B aunque no estn relacionadas con ninguna fila de la tabla A. <BR>





##### FULL OUTER JOIN
Es como la suma de las dos anteriores. Queremos tanto las filas de la A como las de B, tanto si hay match como si no.<BR>
