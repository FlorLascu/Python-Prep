##### Drilling Down vs. Drilling Up <br>


*   Las limitaciones de *Group By* es que solo podemos utilizar las Columnas con Funciones de Agregacion.<br>
*   Mientras que con las funciones ventana, podemos utilizar todas las columnas sin limitaciones<br>
*   Las funciones ventana nos permiten realizar calculos sin perder informacion como con las funciones de **AGREGACION**<br>

#### Window Frame Preceding

*   **UNBOUNDED PRECEDING** specifies that the window starts at the first row of the partition
*   **PRECEDING** indicates the number of rows that precede the current row
*   **CURRENT ROW** that the window starts at the current row

#### Window Frame Following


*   **UNBOUNDED FOLLOWING** specifies that the window ends at the last row of the partition
*   **FOLLOWING** indicates the number of rows that follow the current row
*   **CURRENT ROW** that the window ends at the current row



#### ROW_NUMBER()
**ROW_NUMBER** is a function in the database language Transact-SQL that assigns a unique sequential number to each row in the result set of a query. It is typically used in conjunction with other ranking functions such as DENSE_RANK, RANK, and NTILE to perform various types of ranking analysis.<br>
The **ROW_NUMBER()** function is a window function in SQL that assigns a unique integer to each row within a partition of a result set. The assigned row numbers are based on the order specified in the ORDER BY clause of the OVER clause. This function is often used in conjunction with window functions to perform ranking and pagination tasks.<br>



*   *PARTITION BY* Divides the result set into partitions to wich ROW_NUMBER() is applied. If you omit this clause, the function treats the whole result set as a single partition  <br>
*   *ORDER BY* Specifies the order of rows within each partition. The ROW_NUMBER() is assigned based on this order.

Some common uses are:

1.  **Ranking ROWS** You can use ROW_NUMBER() to assign a rank to each row based on a specific order.<br>
For example, ranking employees by their salary within each department.<br>



2.  **PAGINATION** The function is commonly used in pagination queries to limit the results to a specific page size. For instance, retrieving rows 11 to 20 of a result set.<BR>



3.  **IDENTIFYING DUPLICATES** You can use ROW_NUMBER() to identify and remove duplicate rows from a result set.<br>

#### RANK


**Rank Function** asignes a rank to a particular column.<br>
The **RANK()** function is a window function that assigns a unique rank to each row within a result set based on the values of one or more columns specified in the ORDER BY clause. The rank is assigned in ascending order, with the smallest values receiving a rank of 1.<br>
Here's the basic syntax of the RANK() function:<br>



-   **PARTITION BY**: Divides the result set into partitions to which the RANK() is applied. <br>
If you omit this clause, the function treats the whole result set as a single partition.<br>

*   **ORDER BY**: Specifies the order of rows within each partition. The **RANK()** is assigned based on this order.<br>
Example:<br>


Common use cases for the RANK() function include:<br>
1.  **RANKING ROWS** to assign a rank to each row based on the values in one or more columns.<BR>
For example, ranking employees based on their salary.<BR>

2.  **Top N-Queries** The **RANK()** function is often used in combination with a WHERE clause to filter for the top N rows within each partition.<br>

<br>
This can be useful, for example, to find the top earners in each department.<br>

3.  **Detecting Duplicates** You can use RANK() to identify duplicate values in a result set.<br>

<br>

This can help you find products with the same price.<br>

#### Keep in mind that RANK() returns the same rank for rows with equal values in the ORDER BY clause, and the next rank is skipped accordingly. If two rows have the same value and receive a rank of 1, the next row will receive a rank of 3 (skipping rank 2). If you want to account for ties and assign the same rank to equal values, you can use the DENSE_RANK() function instead.

#### CTE - Common Table Expression<br>
CTE is a named temporary result set defined within the scope of a SELECT, INSERT, UPDATE, or DELETE statement in SQL. <br> 
CTEs provide a way to create complex queries and make SQL code more readable and maintainable.<br>
<br>



Key Components:<br>
1.  **WITH**
Specifies the Start of the CTE Definition 
2.  **CTE_Name** the name to the CTE
3.  **(Column1,Column2,...)** Optional column list that define the column names for the CTE.<br>
4.  **AS** indicates the start of the CTE query<br>
5.  The main query: This is the Select,INSERT,UPDATE, or DELETE statement that references the CTE<br>
<br>

Common use cases for CTEs include:<br>

1. **Recursive Queries**: CTEs can be used for recursive queries, allowing you to work with hierarchical data.<br>

2. **Code Reusability**: CTEs allow you to define complex subqueries and reuse them in the main query or subsequent CTEs.<br>

3. **Readability**: CTEs improve the readability of SQL code by breaking down complex logic into named, modular components.<br>

4. **Window Functions**: CTEs are often used with window functions for tasks such as ranking, partitioning, and aggregating data.<br>
