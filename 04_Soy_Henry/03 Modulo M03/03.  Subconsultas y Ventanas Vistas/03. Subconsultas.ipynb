### Subconsultas
Que son las subconsultas? <br>
Una consulta secundaria es una consulta de seleccion que esta contenida (anidada) dentro de otra consulta.<br>

La consulta de selección interna generalmente se usa para determinar los resultados de la consulta de selección externa.<br>
Consiste en utilizar los resultados de una consulta dentro de otra, que se considera la principal.<br>

Una subconsulta tiene la misma sintaxis que una sentencia SELECT exceptuando que aparece encerrada entre paréntesis, no puede contener la cláusula ORDER BY, ni puede ser la UNIONde varias sentencias SELECT.<br>
<br>
SELECT -> FROM -> WHERE -> GROUP BY -> HAVING -> ORDER BY
<br>
Si lo que retorna la subconsulta es una Lista podemos utilizar los operadores:<br>
Listas --> IN , NOT IN<br>
Valor unico --> = , !=, <> , < , ><br>

<br>

Para eso utilizamos una sub consulta, asi nos trae el registro indicado sin manipularlo adicionalmente<br>
