## Funciones Ventana
Puede entenderse como un conjunto de registros y una funcin que se ejecuta sobre los mismos y cumple determinadas condiciones. <br> 
Para cada registro se debe ejecutar una funcin en esta ventana<br>
Las funciones ventana evita los JOIN de una Tabla consigo misma.<br>




La funcion Ventana se puede descomponer en las siguientes partes:<br>
* **OVER**: Define una ventana o conjunto de filas que debe utilizar la funcion ventana, incluyendo cualquier orden.<br>
    Se pueden utilizar multiples clausulas OVER en una sola consulta, cada una con su propio particionamiento y ordenacion si es necesario<br>
* **Clausula de particion (PARTITION BY)** Las ventanas se agrupan segn esos campos y las funciones de ventana se ejecutan en diferentes grupos.<br>
* **Orden por Clausula (ORDER BY)**:  Segn los campos a ordenar, la funcin de ventana enumerar los registros segn el orden de clasificacin.
 Se puede utilizar junto con la clusula de particin o solo.<br>

* **Clausula de Marco**: EL marco es un subconjunto de particion actual. La clausula se usa para definir las reglas del subconjunto y generalmente se usa como una ventana deslizante.<br>
    * **UNBOUNDED** significa ir todo el camino al limite de la direccion especificada por PRECEDING O FOLLOWING (comienzo o final)
    * **CURRENT ROW** indica inicio o final en la fila actual en la particion <br>
    * **ROWS BETWEEN** permite definir un rango de filas entre dos puntos <br>
