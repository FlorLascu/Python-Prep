### Distribuciones de Probabilidad <br>
Cuando a todos los posibles valores numéricos de una variable aleatoria se le asignan valores de probabilidad<br>
Ya sea mediante:<br>
*  un listado<br>
*  o una *Función matemática* <br> 
el resultado es una distribución de probabilidad. <br>
La suma de las probabilidades de todos los resultados numéricos posibles debe ser igual a 1.<br>
**Una función de probabilidad puede asignar valores de probabilidad a cada estado del espacio muestral.**

#### Características de las Distribuciones de Probabilidad <br>
Las características generales de las distribuciones de probabilidad difieren según el tipo de variable que se encuentre bajo estudio:<br>
1. aleatoria, <br>
2. discreta<br>
3.  o continua<br>
Si la *variable aleatoria es continua*, no pueden listarse todos los posibles valores de la variable <br>
Por esto, las  probabilidades que se determinan por medio de una función matemática son gráficamente representadas por una **función de densidad de probabilidad** , o curva de probabilidad.<br>

### Variable Discreta<br>
- Pueden tomar solamente algunos valores dentro de un intervalo definido<br>
- Las probabilidades se representan con  pi o p(xi).<br>
- El gráfico de la distribución de probabilidad se denomina gráfico de bastones.<br>

* La condición de cierre se verifica realizando la sumatoria de las probabilidades p = 1.<br>

* Las probabilidades se calculan mediante la aplicación de las reglas provenientes de la teoría clásica de probabilidad así como de fórmulas específicas.<br>

#### Ejemplo<br>
Una novela ha tenido un gran éxito, y se estima que el 80% de un grupo de lectores ya la han leído. <br>
En un grupo de 4 amigos aficionados a la lectura:<br>
-- ¿Cuál es la probabilidad de que en el grupo hayan leído la novela 2 personas?<br>
#### Caso binario<br>
Haber leído la novela - No haber leído la novela.<br>


```python
from math import factorial
# n numero de amigos total n=4
# x es el numero de eventos favorables = a  2 amigos haber leido el libro => x =2
# p es la probabilidad de exito donde exito es haber leido el libro => p = 0.8 
# q es la probabilidad de fracaso, es decir, no haber leido el libro q = 1 - p = 0.2
def dist_binomial(n,k,p):
    q = 1 -p
    binomial = (factorial(n) / (factorial(k)*factorial((n-k))))* pow(p,k) * pow(q,(n-k))
    return round(binomial,4) 
dist_binomial(4,2,0.8)
```

```python
dist_binomial(20,6,0.25)
```

25% of all students enrolled in high school XYZ are taking Algebra. 30 students are chosen at random.
a. Find the probability that exactly 7 students out of the 30 chosen are taking algebra.

```python
p7 = dist_binomial(30,7,0.25)
p7
```

b. What is the probability that fewer than 5 students out of the 30 who are selected are taking algebra?

```python
# P(X<5) = P(X=0) + P(X=1) + P(X=2) + P(X=3) + P(X=4)
p0 = dist_binomial(30,0,0.25)
p1 = dist_binomial(30,1,0.25)
p2 = dist_binomial(30,2,0.25)
p3 = dist_binomial(30,3,0.25)
p4 = dist_binomial(30,4,0.25)
print(p0,p1,p2,p3,p4)
print(p0+p1+p2+p3+p4)
```

#### Distribución Hipergeométrica<br>
- Modela el número de eventos en una muestra de tamaño fijo cuando usted conoce el número total de elementos en la población de la cual proviene la muestra.<br> 
- Cada elemento de la muestra tiene dos resultados posibles (es un evento o un no evento). <br> 
- Las muestras *no tienen reemplazo*, por lo que cada elemento de la muestra es diferente.<br>
- Cuando se elige un elemento de la población, no se puede volver a elegir.<br>
- Por lo tanto, la probabilidad de que un elemento sea seleccionado aumenta con cada ensayo, presuponiendo que aún no haya sido seleccionado.<br>

##### Esta estrechamente relacionada con la probabilidad BINOMIAL. La diferencia entre ambas está en la independencia de los intentos y en que la probabilidad de éxito cambia de un a otro.<br>
El resultado de una observación es afectado por los resultados de las observaciones previas, por tanto las probabilidades son condicionales.<br>  
<br>


```python
from math import factorial 
def dist_hipergeometrica(N,n,X,x):
    Xx = factorial(X)/(factorial(x)*factorial((X-x)))
    NX_nx = factorial(N-X)/(factorial(n-x)*factorial((N-X)-(n-x)))
    Nn = factorial(N)/(factorial(n)*factorial(N-n))
    prob_hipergeom = round((Xx * NX_nx)/Nn,2)
    return prob_hipergeom
```

Ejemplo, Una empresa que importa los autos que vende una concesionaria, desea hacer una encuesta de satisfacción a los compradores de estos autos. De una muestra de 80 autos, 30 son importados.<br>
Si se seleccionan 9 clientes.<br> 
¿Cuál es la probabilidad de que haya 2 que compraron autos importados?.<br>

```python
# N = 80 autos
# X = 30
# n = 9
# x = 2
prob_importado_2 = dist_hipergeometrica(80,9,30,2)
prob_importado_2
```

#### Distribución de Bernoulli<br>
Only two possible outcomes = Fail or Success <br>
p = Success<br>
q = Fail = 1 - p<br>

Por ejemplo, en el campo de la calidad, un producto se puede clasificar como bueno o malo.<br>
Las variables de Bernoulli pueden tomar dos valores numéricos, 0 y 1, donde 1 corresponde a un evento y 0 corresponde a un no evento. <br> 
Una variable aleatoria X sigue una distribución de Bernoulli si: <br>
-  P(X = 1) = p <br>
- y P(X = 0) = 1 – p<br>
 donde p es la probabilidad de ocurrencia del evento<br>

 La distribución de Bernoulli es una distribución discreta que está relacionada con muchas distribuciones, tales como la distribución binomial, geométrica y binomial negativa. La distribución de Bernoulli representa el resultado de 1 ensayo. Las secuencias de ensayos de Bernoulli independientes generan las demás distribuciones: la distribución binomial modela el número de éxitos en n ensayos, la distribución geométrica modela el número de fallas antes del primer éxito y la distribución binomial negativa modela el número de fallas antes del éxito xésimo<br>



#### Distribución Poisson<br>
Utilice la distribución de Poisson para describir el número de veces que un evento ocurre en un espacio finito de observación. <br>
La variable aleatoria es el número de veces que ocurre un evento en un  intervalo de tiempo, distancia, área, volumen u otra similar.<br>
<br>
**Por ejemplo**, una distribución de Poisson puede describir el número de defectos en el sistema mecánico de un avión o el número de llamadas a un centro de llamadas en una hora.<br> 
 - La distribución de Poisson se utiliza con frecuencia en el control de calidad, los estudios de fiabilidad/supervivencia y los seguros.<br>
<br>
Una variable sigue una distribución de Poisson si se cumplen las siguientes condiciones:<br>
    - Los datos son conteos de eventos (enteros no negativos, sin límite superior).<br>
    - Todos los eventos son independientes.<br>
    - La tasa promedio no cambia durante el período de interés.<br><br>
    

##### ¿Qué es tasa de ocurrencia?<br>
La tasa de ocurrencia es igual a la media (λ) dividida entre la dimensión del espacio de observación.<br>
Es útil para comparar conteos de Poisson recolectados en diferentes espacios de observación.<br>
** Por ejemplo** , la central telefónica A recibe 50 llamadas telefónicas en 5 horas y la central telefónica B recibe 80 llamadas en 10 horas.<br> Usted no puede comparar directamente estos valores, porque sus espacios de observación son diferentes.<br>
Debe calcular la *tasa de ocurrencia* para comparar estos conteos.<br>
*  La tasa de la central telefónica A es (50 llamadas / 5 horas) = 10 llamadas/hora.<br>
*  La tasa de la central telefónica B es (80 llamadas / 10 horas) = 8 llamadas/hora.<br>
##### Diferencias entre la distribución de Poisson y la distribución binomial
La distribución de Poisson es similar a la distribución binomial porque ambas modelan conteos de eventos. Sin embargo, dentro de su espacio de observación finito, la distribución de Poisson no establece un límite superior a este conteo: una central telefónica podría recibir un número ilimitado de llamadas en un día y no violar los requisitos de la distribución de Poisson. En cambio, la distribución binomial establece un límite superior en el conteo: el número de eventos que usted observa no puede exceder el número de ensayos que realiza.<br>

```python
from math import e,factorial
def dist_poisson(lambda_np,x):
    prob = (pow(lambda_np,x)/factorial(x))*(pow(e,-lambda_np))
    return round(prob,4)
```

A small business receives, on average, 12 customers per day. What is the probability that the business will receive exactly 8 customers in one day?

```python
dist_poisson(12,8)
```

```python
dist_poisson(8,7)
```

La distribución normal es una distribución continua que se especifica por la media (μ) y la desviación estándar (σ). La media es el pico o centro de la curva en forma de campana. La desviación estándar determina la dispersión de la distribución.<br>
