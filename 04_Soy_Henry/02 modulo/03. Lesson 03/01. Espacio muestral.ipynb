#### 2. Excluyentes<br>
Cuando la ocurrencia de uno cualquiera de ellos imposibilita la ocurrencia de los otros<br>
 - Ejemplo de *excluyente*, es la moneda. O sale cara o sale ceca<br>
 - Tirar dos veces la misma moneda, como sale la moneda (cara o cruz) es **excluyente** pero cada evento que es tirar la moneda son *independientes*<br>
##### Ejemplo:
Auto importado o nacional<br>
Son sucesos exhaustivos. O es nacional o es importado. No puede ser ambos <br>

##### Dos sucesos son *compatibles* cuando pueden ocurrir al mismo tiempo.<br>
Ejemplo: nacional o mas de 40 anos. Esta definicion no indica que estos sucesos deban necesariamente ocurrir en forma conjunta.<br>

## Eventos Condicionales<br>
Dos sucesos o eventos son condicionales cuando la ocurrencia o no ocurrencia de un suceso o evento afecta la probabilidad de ocurrencia del otro<br>

<br>
<br>

### Probabilidad condicional<br>
La notacion indica que se está considerando la probabilidad del evento A dada la condición de que el evento B ha ocurrido. <br>
Por lo tanto, la notación P(A/B) se lee *la probabilidad de A dado B* <br>
Que los eventos A y B ocurren simultaneamente<br>
<br>
- Cuando para dos sucesos o eventos cualesquiera A y B la P(A/B) = P(A) => ambos sucesos son independientes <br>
- En este caso también ocurre que P(B/A) = P(B).<br>
