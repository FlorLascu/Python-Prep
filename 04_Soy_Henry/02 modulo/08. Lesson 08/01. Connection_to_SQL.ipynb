```python
cursor = conexion.cursor()
```

```python
# Subir datos desde variables
nombre = 'Gonzalo'
edad = 34
cursor.execute("INSERT INTO test.persona (Nombre,Edad) VALUES(%s,%s)",(nombre,edad))
conexion.commit()
```

```python
# para insertar masivamente valores o registros
# executemany(lista de tuplas)
personas = [('Laura',35),
            ('Jose',25),
            ('Miguel',38),
            ('Leopoldo',28)]
cursor.executemany("INSERT INTO test.persona (Nombre,Edad) VALUES(%s,%s)",personas)
conexion.commit()
# sin el conexion.commit() no se finaliza la ejecucion de los values
```

#### Write Pandas DataFrame into a MySQL Database Table
Invoke to_sql() method on the pandas dataframe instance and specify the table name and database connection.<br>
This creates a table in MySQL database server and populates it with the data from the pandas dataframe

```python
tableName = "ofertaGastro"
dataFrame = ofertaGastro_csv.copy()
sqlEngine = create_engine('mysql+pymysql://root:root1234@127.0.0.1/test',pool_recycle=3600)
dbConnection = sqlEngine.connect()
try:
    frame = dataFrame.to_sql(tableName,dbConnection,if_exists='fail');
except ValueError as vx:
    print(vx)
except Exception as ex:
    print(ex)
else:
    print('Table %s created Successfully.'%tableName);
finally:
    dbConnection.close()
```
