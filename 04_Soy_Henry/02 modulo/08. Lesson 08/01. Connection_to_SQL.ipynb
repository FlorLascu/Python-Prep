```python
cursor.execute(' SELECT * FROM alumno')
alumnos = cursor.fetchmany(5)
for alumno in alumnos:
    print(alumno)
```

```python
# Cada registro se guarda en una tupla. Igual que en MySQL, lo cargamos como una tupla.
# Entonces lo podemos manipular como tuplas en python con todas las sentencias de tuplas.
# Que conviene trabajar en Python solo los registros importantes o que queres visualizar o bajar todo en Python?
# con esta sintaxis se pueden realizar todos los queries igual tal cual que en MySQL
cursor.execute('''SELECT IdAlumno,Nombre,Apellido
               FROM alumno
               LIMIT 5;''')
# alumnos = cursor.fetchall()
for alumno in cursor:
    print(alumno)
```

```python
# Establecer la conexion
# Host
# Puerto
# User
conexion = pymysql.connect(
                host='127.0.0.1', #'tambien puedo poner localhost y tambien funciona' 
                user='root',
                password='root1234',
                database='test',
                port=3306) # 'No hace falta aclararlo, toma el que sea por default, pero si lo aclaro y pongo el default va bien' 
```

```python
cursor = conexion.cursor()
```

```python
# Subir datos desde variables
nombre = 'Gonzalo'
edad = 34
cursor.execute("INSERT INTO test.persona (Nombre,Edad) VALUES(%s,%s)",(nombre,edad))
conexion.commit()
```

```python
# para insertar masivamente valores o registros
# executemany(lista de tuplas)
personas = [('Laura',35),
            ('Jose',25),
            ('Miguel',38),
            ('Leopoldo',28)]
cursor.executemany("INSERT INTO test.persona (Nombre,Edad) VALUES(%s,%s)",personas)
conexion.commit()
# sin el conexion.commit() no se finaliza la ejecucion de los values
```

```python
conexion.close()    # siempre cerrar la conexion.
```
