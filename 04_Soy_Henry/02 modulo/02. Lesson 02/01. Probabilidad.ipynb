### Events can be:<br>
#### 1. Independent<br>
Each event is not afected by other events (dice outcome)<br>

    Example: You toss a coin three times and it comes up "Heads" each time ...<br>
    What is the chance that the next toss will also be a "Head"? <br>

    The chance is simply 1/2, or 50%,<br>
    just like ANY OTHER toss of the coin.<br>

What it did in the past will not affect the current toss!<br>

##### When two events are independent, meaning that the occurrence of one event does not affect the occurrence of the other, you can find the probability of both events happening by multiplying their individual probabilities.<br>

#### 2. Dependent<br>
        Example: Drawing 2 Cards from a Deck<br>
        After taking one card from the deck there are less cards available, so the probabilities change!<br>
        <br>
        <br>
        Let's look at the chances of getting a King.<br>
        <br>
        For the 1st card the chance of drawing a King is 4 out of 52<br>

        But for the 2nd card:<br>
        <br>
        If the 1st card was a King, then the 2nd card is less likely to be a King, as only 3 of the 51 cards left are Kings.<br>
        If the 1st card was not a King, then the 2nd card is slightly more likely to be a King, as 4 of the 51 cards left are King.<br>
        <br>
        Also called *Conditional*, where an event is affected by other events
#### 3. Mutually Exclusive<br>
    Events can't happen at the same time<br>
    It is either one or the other, but not both<br>
    <br>
    Examples:<br>
    <br>
    Turning left or right are Mutually Exclusive (you can't do both at the same time)<br>
    Heads and Tails are Mutually Exclusive<br>
    Kings and Aces are Mutually Exclusive<br>
    What isn't Mutually Exclusive<br>
    <br>
    Kings and Hearts are not Mutually Exclusive, because we can have a King of Hearts!<br>
