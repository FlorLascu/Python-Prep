# Mediana <br>

Mide la observacion central del conjunto.
* Para hallar la mediana de un conjunto de datos, primerso se organizan los datos en orden descendente o ascendente
 * El elemento que esta mas al centro del conjunto de numeros, la mitad de los elementos estan por arriba de este punto y la otra mitad esta por debajo
 * Si el conjunto de datos contiene un numero impar de elementos, el del medio en un arreglo es la mediana<br>
 - Si hay un numero par de observaciones, la mediana es el proemdio de los dos elementos del medio<br>

 Ventajas<br>
 - No se ve afectada por los valores extremos <br>
- Es facil de entender y se calcula a partr de cualquier tipo de datos <br>
 - La podemos encontrar incluso cuando nuestros datos descripticion cualitativas, en lugar de numeros<br>

 Desventajas<br>

 - Datos ordenados antes de cualquier calculo <br>
 - Ciertos procedimientos estadisticos que utilizan la mediada  son mas complejos que aquellos que utilizan la media<br>




# Moda <br>

La moda es el valor que mas se repite en el conjunto de datos.<br>


como se encuentra????<br>
* se toma la frecuencia de las observacinoes y la de mayor frecuencia es la moda<br>
<br>
Ventaja<br>
* se puede utilizar como una posicion central para datos tanto cualitativos como cuantitativos <br>
* La mediana, los valores extremos no afectan indebidamente a la moda<br>
* La podemos utilizar aun cuando una o mas clases sean de extremo abierto<br> 
<br>
Desventajas<br>

* No se usa tan amenudo como medida de tendencia central<br>
* A veces no existe un valor<br>
* Puede haber mas de un valor modal en un conjunto de datos => bimodal<br>




<br>

<br>

Porque? Porque los valores extremos afectan a la media <br>
La moda siempre va a estar en el punto mas alto de la distribucion.
