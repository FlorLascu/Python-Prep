```python
hospitales_csv[['Point', 'longitude','latitude']] = hospitales_csv.WKT.str.split(" ",expand=True)
# hospitales_csv.head(5)
```

```python
latitud =[float(elemento.strip(')')) for elemento in hospitales_csv['latitude']]
hospitales_csv['latitude'] = latitud
```

```python
longitud = [float(elemento.strip('()')) for elemento in hospitales_csv['longitude']]
hospitales_csv['longitude'] = longitud
```

```python
# hospitales_csv.head(5)
```

```python
hospitales_csv_copia = hospitales_csv.rename(columns = {'DOM_GEO':'name','NOM_MAP':'label'})
hospitales_csv_copia.drop(['WKT', 'ID', 'NOMBRE', 'OBJETO', 'CALLE', 'ALTURA',
        'DOM_NORMA', 'TELEFONO', 'GUARDIA', 'FAX', 'WEB',
       'COD_POSTAL', 'TIPO', 'TIPO_ESPEC', 'MOD_AT_1', 'MOD_AT_2',
       'DEPEND_ADM', 'DIRECTOR', 'Point'],axis=1,inplace=True)
hospitales_csv_copia
```

```python
hospitales_csv_copia = hospitales_csv_copia[[ 'latitude', 'longitude', 'name','label']]

hospitales_csv_copia.head(2)
```

```python
# ahora lo exporto a un csv file
hospitales_csv_copia.to_csv(r'C:\Users\Florencia\Desktop\Python-Prep\Carpeta Nueva\modulo_01\lesson_04\hospitales_csv_copia.csv',sep=',',index=False)
```
