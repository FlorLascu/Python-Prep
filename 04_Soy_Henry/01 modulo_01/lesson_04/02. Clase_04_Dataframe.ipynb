DataFrame

* Un DataFrame contiene dos índices, uno para las filas y otro para las columnas, y se puede acceder a sus elementos mediante los nombres de las filas y las columnas. 
Es un conjunto de datos estructurado en forma de Tabla y las filas son los registros que pueden contenter los distintos tipos de datos
* Tabla de excel
* las puedo crear a partir:
    - de un diccionario (Un diccionario de listas)
    - de lista de listas
    - de un archivo .csv (levantamos datos de .txt)

```python
# Crear un DataFrame a partir de un diccionario con el constuctor de Diccionarios dict de Pyhton
# el Diccionario es un diccionario de listas. Claves : [lista]
# DataFrame(data=diccionario, index=filas, columns=columnas, dtype=tipos)

notas = pd.DataFrame(dict(juan=[6,3,9],alessia=[7,8,7],jimena=[9,10,6]))

# ambas filas y columnas pueden tener nombres. En el caso del Diccionario, los nombres de las columnas son las claves del dicc.
# si no nombramos a las filas, los indices toman por default un id de numero.

notas
```

```python
 # Creamos un DF a partir de una lista de listas:
 # DataFrame(data=listas, index=filas, columns=columnas, dtype=tipos) : Devuelve un objeto del tipo DataFrame 
 # cuyas columnas son los valores de las listas de la lista listas, los nombres de filas indicados en la lista filas,
 # los nombres de columnas indicados en la lista columnas y los tipos indicados en la lista tipos.
 
data = [['DS', 'Linked_list', 10], ['DS', 'Stack', 9], ['DS', 'Queue', 7],['Algo', 'Greedy', 8], ['Algo', 'DP', 6], ['Algo', 'BackTrack', 5], ] 
print(data)
df1 = pd.DataFrame(data,columns=['Category', 'Name', 'Marks'])
print(df1)

# los nombres de filas indicados en la lista filas

df1 = pd.DataFrame(data,index=['cat 01','cat 02','cat 03','cat 04','cat 05','cat 06'],columns=['Category', 'Name', 'Marks'])
print(df1)
```

```python
# Creación de un DataFrame a partir de una lista de diccionarios
# DataFrame(data=diccionarios, index=filas, columns=columnas, dtype=tipos) 
# las claves del diccionarios son los headers de la tabla
# y los values del diccionario forman los registros/filas del DF

df3 = pd.DataFrame([{'Nombre':'María', 'Edad':18}, {'Nombre':'Luis', 'Edad':22}, {'Nombre':'Carmen'}])
print(df3)
```

```python
# Convertir un DataFrame a formato largo 
# de formato ancho a formato largo (columnas a filas)
# df.melt(id_vars=id-columnas, value_vars=columnas, var_name=nombre-columnas, var_value=nombre-valores)
# id_vars = son las columnas que permanecen
# value_vars = son los valores que se merge o que se convierten en datos de la nueva columna
# var_name = es el nombre de la variable donde se guardan los headers por fila = merge las columnas deseadas
# var_value = es el nombre de la columna que aceptara los valores que se corresponden al merge 
```

```python
import pandas as pd
datos = {'nombre':['María', 'Luis', 'Carmen'], 'edad':[18, 22, 20],'Matemáticas':[8.5, 7, 3.5],'Economía':[8, 6.5, 5],
 'Programación':[6.5, 4, 9]}
df5 = pd.DataFrame(datos)
print("este es el formato ancho de DataFrame"+ '\n' + '\n' + f'{df5}')
```

```python
df6 = df5.melt(id_vars=['nombre', 'edad'], var_name='asignatura', value_name='nota')
print("este es el formato largo de DataFrame"+ '\n' '\n'+f'{df6}')
```

```python
# Convertir un DataFrame a largo a  formato ancho 
# df.pivot(index=filas, columns=columna, values=valores) :
# se crean tantas columnas como agrupacion de datos haya
# y los valores de las nuevas columnas seran los datos de las filas de la columna elegida
# values = valores => los datos que se guardaran en las nuevas columnas 

print(df6.pivot(index='nombre', columns='asignatura', values='nota'))
```
