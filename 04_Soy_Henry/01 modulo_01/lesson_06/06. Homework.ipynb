```python
def hash_function(key):
    return sum(index * ord(character) for index, character in enumerate(repr(key), start=1))
```

```python
# aplicamos la funcion hash a la columina cod_pais para crear la columna clave_hash
emisiones_csv['Clave_Hash'] = emisiones_csv['cod_pais'].apply(hash_function)
emisiones_csv.head(4)
```

```python
emisiones_csv.nunique() # muestra que para codigo pais y pais hay 199 registros unicos.
# pero la clave hash tiene 108, entonces significa que esta habiendo un codigo pais y nombre de pais que tiene el mismo hash.
# por lo tanto, hay que mejorar la funcion hash o POLITICA DE COLISIONES
```

```python
emisiones_csv_copia = emisiones_csv.copy()
emisiones_csv_copia['Clave_Hash'] = emisiones_csv_copia.apply(lambda row: hash_function(row['Cod_pais']+row['Nom_pais']+row['Región']), axis=1)
emisiones_csv_copia.info()
```
