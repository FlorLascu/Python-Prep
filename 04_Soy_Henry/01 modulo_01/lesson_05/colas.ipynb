```python
class Cola:
    def __init__(self):
        self.cola = []
        self.size = 0
    
    #Esta vacia la cola?       
    def empty(self):
        return len(self.cola) == 0
    
    # agregar datos a la cola
    def push(self,dato):
        self.cola += [dato]
        self.size += 1
    
    # eliminar un elemento de la cola 
    def pop(self):
        if self.empty():
            print('la cola esta vacia')
        else:
            self.cola = [self.cola[i] for i in range(1,self.size)]
            # aca reescribimos la cola, solo que al empezar de 1 en vez de 0, 
            # se elimina solo el elemento primero, que estaba en la posicion 0
            # otra forma seria utilizar .pop(0)
            self.size -= 1
    # mostrar la cola
    
    def show(self):
        n =self.size -1  # le resto 1 porque las listas empiezan de cero y size suma de a uno desde 1
        while n > -1:
            print(f'[{n}] => {self.cola[n]}')
            n -=1
    
    # mostrar el primer elemento de la cola
    def front(self): 
        if self.empty(): print("cola vacia")
        else: print("primer dato: ", self.cola[0])
try:
    if __name__ == "__main__":
        opcion = 0
        cola = Cola()
        while opcion != 6:
            print("--- Colas---\n 1. Agregar dato \n 2. Eliminar dato \n 3. Esta vacia la cola? \n 4. Mostrar cola \n 5. mostrar primer dato \n 6. Salir")
            opcion = int(input('Ingresa un numero'))
        if opcion == 1:
            dato = input('ingresa un numero ')
            cola.push(dato)
        elif opcion == 2:
            cola.pop()
        elif opcion == 3:
            print('SI' if cola.empty() else 'NO')
        elif opcion == 4:
            print(cola.show())
        elif opcion == 5:
            print(cola.front())
        elif opcion == 6:
            print('\n --- Sesion terminada---')
        else: print('ingresaste una opcion erronea, vuelve a intentarlo')
except: Exception         
```
